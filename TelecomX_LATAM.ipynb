#📌 Extracción

```python
import pandas as pd
url_api = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"
datos = pd.read_json(url_api)
datos.head()
```

#🔧 Transformación

```python
datos.info()
```

```python
datos.dtypes
```

```python
datos_normalizados = pd.json_normalize(datos.to_dict('records'))
datos_normalizados.head()
```

```python
datos_normalizados
```

```python
datos_normalizados.info()
```

```python
datos_normalizados.columns = datos_normalizados.columns.str.replace('customer.', '')
datos_normalizados.columns
```

```python
datos_normalizados.isnull().sum()
```

```python
datos_normalizados['account.Charges.Total'] = pd.to_numeric(datos_normalizados['account.Charges.Total'], errors='coerce')
datos_normalizados['account.Charges.Total']
```

##Datos duplicados

```python
datos_normalizados.duplicated().sum()
```

##Correccion de espacios Invisibles

```python
for col in datos_normalizados.select_dtypes(include='object'):
  datos_normalizados[col]  = datos_normalizados[col].str.strip()
```

##Cuentas Diarias

```python
  datos_normalizados['Cuentas_Diarias'] = datos_normalizados['account.Charges.Monthly'] / 30
  datos_normalizados[['account.Charges.Monthly','Cuentas_Diarias']].head()
```

##Transformacion de Datos

```python
for col in datos_normalizados.select_dtypes(include='object'):
  datos_normalizados[col] = datos_normalizados[col].str.lower()
```

##Datos a Binario

```python
Col_binarias = [
    'Churn',
    'Partner',
    'Dependents',
    'phone.PhoneService',
    'phone.MultipleLines',
    'internet.DeviceProtection',
    'internet.StreamingTV',
    'internet.StreamingMovies',
    'internet.TechSupport',
    'account.PaperlessBilling'
 ]

for col in Col_binarias:
    datos_normalizados[col] = (
       datos_normalizados[col]
       .map({'yes': 1, 'no': 0})
       .astype('Int64')
      )
datos_normalizados.head()

```

##Traduciendo columnas

```python
datos_normalizados = datos_normalizados.rename(columns={
'customerID': 'ID_Cliente',
'Churn': 'Abandono',
'gender': 'Género',
'SeniorCitizen': 'Adulto_Mayor',
'Partner': 'Pareja',
'Dependents': 'Dependientes',
'tenure': 'Antigüedad_Meses',
'phone.PhoneService': 'Teléfono_Servicio',
'phone.MultipleLines': 'Teléfono_Múltiples_Líneas',
'internet.InternetService': 'Internet_Tipo_Servicio',
'internet.OnlineSecurity': 'Internet_Seguridad_En_Línea',
'internet.OnlineBackup': 'Internet_Respaldo_En_Línea',
'internet.DeviceProtection': 'Internet_Protección_Dispositivo',
'internet.TechSupport': 'Internet_Soporte_Técnico',
'internet.StreamingTV': 'Internet_TV_Streaming',
'internet.StreamingMovies': 'Internet_Películas_Streaming',
'account.Contract': 'Cuenta_Tipo_Contrato',
'account.PaperlessBilling': 'Cuenta_Facturación_Sin_Papel',
'account.PaymentMethod': 'Cuenta_Método_Pago',
'account.Charges.Monthly': 'Cuenta_Cargo_Mensual',
'account.Charges.Total': 'Cuenta_Cargo_Total',
'Cuentas_Diarias': 'Cuentas_Diarias'
})
datos_normalizados.columns
```

El objetivo de este análisis es estudiar la evasión de clientes en una empresa de telecomunicaciones. La evasión ocurre cuando un cliente decide cancelar el servicio.

Este estudio busca identificar patrones que expliquen este comportamiento.

- Eliminación de valores nulos.

##Análisis Exploratorio


### Distribución de evasión

Se observó que la mayoría de clientes permanece en la empresa, pero existe un porcentaje importante que cancela.



### Variables categóricas

Se encontró que los contratos mensuales tienen mayor nivel de evasión.



---

###  Variables numéricas

Los clientes con menor antigüedad y cargos mensuales más altos presentan mayor tendencia a cancelar.

## Conclusiones

- El tipo de contrato influye en la evasión.
- Los clientes nuevos tienen mayor riesgo de cancelar.
- El precio puede afectar la permanencia.
